5) Write a function color_mixer(primary1, primary2) that takes two color strings (e.g., "red", "blue") and returns a new color string describing the mix (e.g., "purple"). For this beginner version, handle only: red+blue -> purple, red+yellow -> orange, blue+yellow -> green. If the combination isnâ€™t in the list, return "unknown color mix".
   - Test: color_mixer("red","blue") -> "purple"

```python
5
def color_mixer(primary1, primary2):
    mix = {
        ("red", "blue"): "purple",
        ("red", "yellow"): "orange",
        ("blue", "yellow"): "green",
        ("blue", "red"): "purple",
        ("yellow", "red"): "orange",
        ("yellow", "blue"): "green",
    }
    key = (primary1, primary2)
    return mix.get(key, "unknown color mix")
Example:
python
color_mixer("red", "blue")  # "purple"
color_mixer("red", "green")
```

8) Write a function is_palindrome(s) that checks whether a string s reads the same forward and backward. Return True or False. Ignore case and spaces for this simple version (e.g., "A man a plan" should be treated as a palindrome).
   - Hint: you can preprocess s by removing spaces and lowering case.
   - Test: is_palindrome("Racecar") -> True, is_palindrome("Hello") -> False

```python
8
def is_palindrome(s):
    cleaned = "".join(ch.lower() for ch in s if ch != " ")
    return cleaned == cleaned[::-1]
Example:
python
is_palindrome("Racecar")  # True
is_palindrome("Hello")
```

9) Write a function multiply_until(n, limit) that starts with a value of 1 and keeps multiplying by n until the product would exceed limit. Return the final product. If limit is less than 1, return 1.
   - Test: multiply_until(3, 20) -> 18 (since 1*3=3, 3*3=9, 9*3=27 would exceed 20)

```python
9
def multiply_until(n, limit):
    product = 1
    if limit < 1:
        return 1
    while product * n <= limit:
        product *= n
    return product
Example:
python
multiply_until(3, 20)  # 18
```

10) Write a function fibonacci(n) that returns the nth Fibonacci number, where fib(0) = 0, fib(1) = 1, and fib(n) = fib(n-1) + fib(n-2) for n >= 2.
    - Test: fibonacci(6) -> 8

```python
10
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b
Example:
python
fibonacci(6)  # 8
```

12) Write a function is_prime(p) that returns True if p is a prime number, otherwise False. Assume p is a positive integer.
    - Hint: check divisibility from 2 up to sqrt(p).
    - Test: is_prime(5) -> True, is_prime(9) -> False

```python
12
import math

def is_prime(p):
    if p <= 1:
        return False
    if p <= 3:
        return True
    if p % 2 == 0 or p % 3 == 0:
        return False
    i = 5
    w = 2
    while i * i <= p:
        if p % i == 0:
            return False
        i += w
        w = 6 - w
    return True
Example:
python
is_prime(5)  # True
is_prime(9)  # False
```


13) Write a function count_vowels(s) that returns the number of vowels (a, e, i, o, u) in the string s. Treat uppercase and lowercase as the same.
    - Test: count_vowels("Hello World") -> 3


```python
13
def count_vowels(s):
    vowels = set("aeiouAEIOU")
    return sum(1 for ch in s if ch in vowels)
Example:
python
count_vowels("Hello World")  # 3
```

    - Test: average_of_list([1, 2, 3, 4]) -> 2.5

```python
16
def average_of_list(nums):
    if not nums:
        return None
    return sum(nums) / len(nums)
Example:
python
average_of_list([1, 2, 3, 4])  # 2.5
```

17) Write a function to_upper_first_word(sentence) that capitalizes only the first character of the sentence and leaves the rest unchanged.
    - Test: to_upper_first_word("hello world") -> "Hello world"

```python
17
def to_upper_first_word(sentence):
    if not sentence:
        return sentence
    return sentence[0].upper() + sentence[1:]
Example:
python
to_upper_first_word("hello world")  # "Hello world"
```


20) Write a function factorial(n) that returns n! for a non-negative integer n. If n is 0, return 1.
    - Test: factorial(5) -> 120


```python
20
def factorial(n):
    if n < 0:
        raise ValueError("n must be non-negative")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
Example:
python
factorial(5)  # 120
```
